```python
# Breast Cancer Classification (scikit-learn)
# ============================================================
# This notebook demonstrates an end-to-end supervised classification workflow
# on the Breast Cancer Wisconsin dataset using scikit-learn.
# We compare three models:
#   1. Logistic Regression (with standardization)
#   2. Random Forest
#   3. MLPClassifier (neural network, with standardization)
#
# We evaluate them on Accuracy, Precision, Recall, F1, and ROC-AUC.
# Finally, we plot ROC curves and visualize the confusion matrix
# for the best-performing model.
# ============================================================

# ------------------------------------------------------------
# Step 1: Imports and warning suppression
# ------------------------------------------------------------
from sklearn.exceptions import ConvergenceWarning
import warnings
warnings.filterwarnings("ignore", category=ConvergenceWarning)  # suppress MLP convergence warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# scikit-learn imports
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, confusion_matrix
)

# ------------------------------------------------------------
# Step 2: Load and inspect the dataset
# ------------------------------------------------------------
data = load_breast_cancer(as_frame=True)
X = data.data
y = data.target

print("Samples:", X.shape[0])
print("Features:", X.shape[1])
print("Classes:", data.target_names.tolist())
print("Class counts:", dict(zip(data.target_names, np.bincount(y))))

# Optional preview of data
display(X.head())

# ------------------------------------------------------------
# Step 3: Split dataset into training and test sets
# ------------------------------------------------------------
# We use a stratified split to preserve class proportions in both sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])

# ------------------------------------------------------------
# Step 4: Build model pipelines
# ------------------------------------------------------------
# We use pipelines to link preprocessing (like scaling) to each model.
models = {
    "LogReg": Pipeline([
        ("scaler", StandardScaler()),  # normalize features for LR
        ("clf", LogisticRegression(max_iter=500, random_state=42))
    ]),
    "RandomForest": Pipeline([
        ("clf", RandomForestClassifier(n_estimators=200, random_state=42))
    ]),
    "MLP": Pipeline([
        ("scaler", StandardScaler()),  # scaling helps neural networks converge
        ("clf", MLPClassifier(hidden_layer_sizes=(32,),
                              max_iter=500, random_state=42))
    ])
}

# ------------------------------------------------------------
# Step 5: Train and evaluate each model
# ------------------------------------------------------------
results = {}
fitted = {}

for name, pipe in models.items():
    print(f"\nTraining {name}...")
    pipe.fit(X_train, y_train)
    fitted[name] = pipe

    # Predictions on the test set
    y_pred = pipe.predict(X_test)

    # Probability scores (for ROC-AUC)
    if hasattr(pipe.named_steps.get("clf", pipe), "predict_proba"):
        y_score = pipe.predict_proba(X_test)[:, 1]
    else:
        y_score = pipe.decision_function(X_test)

    # Compute metrics
    results[name] = {
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred),
        "recall": recall_score(y_test, y_pred),
        "f1": f1_score(y_test, y_pred),
        "roc_auc": roc_auc_score(y_test, y_score)
    }

# Create a DataFrame of results, sorted by ROC-AUC
df_results = pd.DataFrame(results).T.sort_values(by="roc_auc", ascending=False)
print("\nEvaluation metrics (sorted by ROC-AUC):")
display(df_results)

# ------------------------------------------------------------
# Step 6: Plot ROC curves
# ------------------------------------------------------------
plt.figure(figsize=(6, 5))
for name, pipe in fitted.items():
    if hasattr(pipe.named_steps.get("clf", pipe), "predict_proba"):
        y_score = pipe.predict_proba(X_test)[:, 1]
    else:
        y_score = pipe.decision_function(X_test)
    fpr, tpr, _ = roc_curve(y_test, y_score)
    plt.plot(fpr, tpr, label=name)

# Add reference diagonal
plt.plot([0, 1], [0, 1], linestyle="--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves - Breast Cancer Classification")
plt.legend()
plt.tight_layout()
plt.show()

# Optional: save ROC curve for GitHub or report
plt.gcf().savefig("roc_curves.png", dpi=200, bbox_inches="tight")

# ------------------------------------------------------------
# Step 7: Confusion matrix for the best model
# ------------------------------------------------------------
best_model_name = df_results.index[0]
best_model = fitted[best_model_name]
y_pred_best = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(5, 4))
plt.imshow(cm, interpolation="nearest")
plt.title(f"Confusion Matrix - {best_model_name}")
plt.colorbar()

tick_marks = np.arange(len(data.target_names))
plt.xticks(tick_marks, data.target_names, rotation=45)
plt.yticks(tick_marks, data.target_names)

# Label each cell
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'), ha="center", va="center")

plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.tight_layout()
plt.show()

# Optional: save confusion matrix figure
plt.gcf().savefig("confusion_matrix.png", dpi=200, bbox_inches="tight")

print(f"\nBest model by ROC-AUC: {best_model_name}")
print("Confusion matrix:\n", cm)
```
